```python
import astropy
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy.cosmology import FlatLambdaCDM
universe = FlatLambdaCDM(H0=70,Om0=1.0)
redshift_array = np.linspace(0,20,250)
comoving_distance = universe.comoving_distance(redshift_array)
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()
ax.set_xlabel("Redshift (Z)")
ax.set_ylabel("Comoving Distance (Mpc)")
ax.set_xlim(0, 20)
ax.set_ylim(0, 7000)
ax.plot(redshift_array, comoving_distance)
```

```python
from astropy.cosmology import WMAP9 as cosmo #to use original hubble constant value 
universe_1 = (H0=cosmo.H(0),Om0=1.0)
universe_1.H0
```

```python
universe_1.Ogamma0 #verified radiation density = 0
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()
ax.set_xlabel("Redshift (Z)")
ax.set_ylabel("Age (Gyr)")
ax.set_xlim(0, 20)
ax.set_ylim(0, 10)
ax.plot(redshift_array, universe_1.age(redshift_array))
```
