As shown in problem 1, $S_k(r) = r$. So, $d_L = r(1+z)$, where r is the comoving distance.

$\textbf{a.)}$ I believe I completed this derivation in the previous problem, when I evaluated the distance to the galaxies as a function of their redshifts. So, to answer this part of the homework I will better explain what I did, and show more steps than I did previously. 

As shown in problem 1 part b, ${\displaystyle E(z)={\sqrt {\Omega _{r}(1+z)^{4}+\Omega _{m}(1+z)^{3}+\Omega _{k}(1+z)^{2}+\Omega _{\Lambda }}}} = \sqrt{\Omega_m(1+z)^3}=\sqrt{(1+z)^3}$, which was also given in lecture.

$\cfrac{dz}{E(z)}$ can be thought of as the ratio of the travel time of a photon across the infintesimal redshift dz and the scale factor at that time. It follows that given a constant speed of light c, this is the same thing as comoving distance because it is simply the proper distance devided by the scale factor. So, finally, the total comoving distance is just integrating this proportionality from 0 to some redshift z, multiplied by $\cfrac{c}{H_0}$, which is the Hubble time multiplied by the speed of light.

This yields the equation I showed in question 1, $r = \frac{c}{H_0} \int_0^z \frac{dz}{(1+z)^{\frac{3}{2}}}$, by using the equation listed above for $E(z)$ in a completely matter universe.

Now, $r = \frac{c}{H_0} \int_0^z \frac{dz}{(1+z)^{\frac{3}{2}}} = \frac{c}{H_0} \big(\int _1^{1+z}\frac{1}{u^{\frac{3}{2}}}du\big)$, where u = 1+z. The inner portion of the integral evaluates to $\left[\cfrac{u^{-\frac{3}{2}+1}}{-\frac{3}{2}+1}\right]^{1+z}_1$ using the power rule.

$\left[\cfrac{u^{-\frac{3}{2}+1}}{-\frac{3}{2}+1}\right]^{1+z}_1$ Now, simplifying the interior yields $ \cfrac{u^{-\frac{1}{2}}}{-\frac{1}{2}} = -\cfrac{2}{u^{\frac{1}{2}}}$, so $\left[\cfrac{u^{-\frac{3}{2}+1}}{-\frac{3}{2}+1}\right]^{1+z}_1 = \left[-\cfrac{2}{u^{\frac{1}{2}}}\right]^{1+z}_1 = \cfrac{-2}{(1+z)^{\frac{1}{2}}} - \cfrac{-2}{1^{\frac{1}{2}}} = 2 - \cfrac{2}{\sqrt{(1+z)}}$ Finally, multiply the inner portion of the integral by the previously mentioned constant $\cfrac{c}{H_0}$, and this yields $\cfrac{c}{H_0}\bigg(2-\cfrac{2}{\sqrt{1+z}}\bigg)$ for the comoving distance as a function of redshift.

Finally, as previously stated, $d_L = r(1+z)$, so $d_L = \left[\cfrac{c}{H_0}\bigg(2-\cfrac{2}{\sqrt{1+z}}\bigg)\right](1+z) =\cfrac{c}{H_0}\left[\cfrac{2\sqrt{1+z}}{\sqrt{1+z}}-\cfrac{2}{\sqrt{1+z}}\right](1+z) = \cfrac{c\frac{2\sqrt{z+1}-2}{\sqrt{z+1}}\left(z+1\right)}{H_0} = \cfrac{c\left(2\sqrt{z+1}-2\right)\sqrt{z+1}}{H_0}$

$\textbf{b.)}$ See below

Note: $c \approx 299792.458 \cfrac{km}{s}$, $H_0 \approx 70 \cfrac{km}{\frac{s}{Mpc}}$, so $d_L$ is in units of Mpc.

```python
import astropy
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()
z = np.linspace(0,3,250)
r = (299792.458/70)*(2-(2/np.sqrt(1+z)))
luminosity_distance = r*(1+z)
angular_diameter_distance = r/(1+z)
ax.set_xlabel("Redshift (Z)")
ax.set_xlim(0, 3)
ax.set_ylabel("Luminosity Distance")
ax.set_ylim(17125)
ax.plot(z, luminosity_distance)
ax.plot(z, angular_diameter_distance)
plt.gca().invert_yaxis() #It was inverted for some reason, so I flipped it here. Not sure why.
```

$\textbf{c.)}$ Angular diameter distance and luminosity difference have different dependencies on redshift, because the surface brightness of an object moving away from an observer is reduced by a factor of $(1+z)^{-4}$, and the angular area goes down by $d_A ^{-2}$. It follow that $d_L = (1+z)^{2}d_a$, so their physical relationship is simply a result of their individual dependencies on redshift, so luminosity distance scales differently than angular diameter distance both by mathematical definition, and because of what each measurement is representing.
